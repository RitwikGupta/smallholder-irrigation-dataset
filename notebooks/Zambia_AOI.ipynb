```python
# create a map of Zambia with the grid cells mapped out
import geopandas as gpd
import matplotlib.pyplot as plt

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
zambia = world[world.name == 'Zambia']

grid = gpd.GeoDataFrame(grid, geometry=gpd.points_from_xy(grid.longitude, grid.latitude))
grid[grid.country == 'Zambia'].plot(ax=zambia.plot(), color='lightgreen', markersize=.1)
```
